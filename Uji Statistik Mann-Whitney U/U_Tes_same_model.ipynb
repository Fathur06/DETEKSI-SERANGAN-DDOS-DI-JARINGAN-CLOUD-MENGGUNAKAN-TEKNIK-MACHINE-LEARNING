# Import Library Mann Whitney

```python
import pandas as pd
from scipy.stats import mannwhitneyu
```

# Load Dataset

```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
```

# RF

# RF BASELINE VS RF CFS

Read Dataset

```python
rf_base = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RF/baseline_rf_results.csv')
rf_cfs  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RF/rf_cfs_results.csv')
```

Mulai Uji

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score', 'Kappa']
alpha = 0.05

print("=== Uji Mann-Whitney: RF_Baseline vs RF_CFS ===\n")

for metric in metrics:
    stat, p = mannwhitneyu(rf_base[metric], rf_cfs[metric], alternative='two-sided')
    print(f"{metric}:")
    print(f"  U-Statistik = {stat:.4f}")
    print(f"  p-value     = {p:.6f}")
    if p < alpha:
        print("  ðŸ‘‰ Perbedaan SIGNIFIKAN (p < 0.05)\n")
    else:
        print("  ðŸ‘‰ Perbedaan TIDAK SIGNIFIKAN (p â‰¥ 0.05)\n")
```
