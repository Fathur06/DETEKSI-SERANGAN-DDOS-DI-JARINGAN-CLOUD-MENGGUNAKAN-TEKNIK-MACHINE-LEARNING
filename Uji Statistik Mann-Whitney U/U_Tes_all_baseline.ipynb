# Boxplot ke-3 Baseline

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load hasil baseline dari ketiga model
rf = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/RF/baseline_rf_results.csv")
knn = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/KNN/baseline_knn_results.csv")
svm = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/SVM/baseline_svm_results.csv")

# Tambahkan label nama model ke masing-masing
rf['model'] = 'Random Forest'
knn['model'] = 'KNN'
svm['model'] = 'SVM'

# Gabungkan jadi satu DataFrame
df = pd.concat([rf, knn, svm], ignore_index=True)

# Ubah ke long format untuk visualisasi seaborn
df_long = pd.melt(df,
                  id_vars=['Fold', 'model'],
                  value_vars=['Accuracy', 'Precision', 'Recall', 'F1_Score', 'Kappa'],
                  var_name='metrik',
                  value_name='nilai')

# Buat boxplot
plt.figure(figsize=(12, 6))
sns.boxplot(x='metrik', y='nilai', hue='model', data=df_long)
plt.title('Perbandingan Performa Baseline (Random Forest vs KNN vs SVM)')
plt.xlabel('Metrik Evaluasi')
plt.ylabel('Nilai Metrik')
plt.legend(title='Model')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# RF BASELINE VS KNN BASELINE

Read Dataset

```python
rf_base = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RF/baseline_rf_results.csv')
knn_base  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/KNN/baseline_knn_results.csv')
```

Mulai Uji

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score', 'Kappa']
alpha = 0.05

print("=== Uji Mann-Whitney: RF_Baseline vs KNN_Baseline ===\n")

for metric in metrics:
    stat, p = mannwhitneyu(rf_base[metric], knn_base[metric], alternative='two-sided')
    print(f"{metric}:")
    print(f"  U-Statistik = {stat:.4f}")
    print(f"  p-value     = {p:.6f}")
    if p < alpha:
        print("   Perbedaan SIGNIFIKAN (p < 0.05)\n")
    else:
        print("   Perbedaan TIDAK SIGNIFIKAN (p  0.05)\n")
```
