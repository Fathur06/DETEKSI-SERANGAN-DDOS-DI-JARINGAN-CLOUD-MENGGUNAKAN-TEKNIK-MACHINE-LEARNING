# 1. RF CFS VS KNN CFS

Read Dataset

```python
rf_cfs = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/RF/rf_cfs_results.csv')
knn_cfs  = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/KNN/knn_cfs_results.csv')
```

Mulai Uji

```python
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score', 'Kappa']
alpha = 0.05

print("=== Uji Mann-Whitney: RF_CFS vs KNN_CFS ===\n")

for metric in metrics:
    stat, p = mannwhitneyu(rf_cfs[metric], knn_cfs[metric], alternative='two-sided')
    print(f"{metric}:")
    print(f"  U-Statistik = {stat:.4f}")
    print(f"  p-value     = {p:.6f}")
    if p < alpha:
        print("  ðŸ‘‰ Perbedaan SIGNIFIKAN (p < 0.05)\n")
    else:
        print("  ðŸ‘‰ Perbedaan TIDAK SIGNIFIKAN (p â‰¥ 0.05)\n")
```
