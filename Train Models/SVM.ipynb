# 3. Pre-processing Data

```python
# Cek nilai unik label
print("Nilai unik pada kolom 'label':", df['label'].unique())

# Encode label: BENIGN = 0, DDOS = 1
df['label'] = df['label'].map({'BENIGN': 0, 'DDOS': 1})

# Pastikan tidak ada nilai NaN setelah encoding
print("Cek nilai NaN setelah encoding:", df['label'].isnull().sum())

# Pisahkan fitur dan label
X = df.drop(columns=['label'])  # Semua kolom kecuali label
y = df['label']

# Normalisasi fitur
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Ubah kembali ke DataFrame untuk memudahkan analisis ke depan
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)

print("Normalisasi selesai. Bentuk data setelah normalisasi:", X_scaled_df.shape)
```

# Train Baseline SVM + Evaluasi

```python
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score
import numpy as np

# Inisialisasi model SVM (kernel RBF)
svm = SVC(kernel='rbf', random_state=42)

# K-Fold Cross Validation
skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

# Simpan hasil evaluasi
accuracy_scores = []
precision_scores = []
recall_scores = []
f1_scores = []
kappa_scores = []

for fold, (train_index, test_index) in enumerate(skf.split(X_scaled_df, y), 1):
    X_train, X_test = X_scaled_df.iloc[train_index], X_scaled_df.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Training SVM
    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)

    # Evaluasi metrik
    accuracy_scores.append(accuracy_score(y_test, y_pred))
    precision_scores.append(precision_score(y_test, y_pred))
    recall_scores.append(recall_score(y_test, y_pred))
    f1_scores.append(f1_score(y_test, y_pred))
    kappa_scores.append(cohen_kappa_score(y_test, y_pred))

    print(f"Fold {fold} selesai")

# Tampilkan rata-rata hasil
print("\n=== Baseline SVM Evaluation (10-Fold CV) ===")
print(f"Accuracy: {np.mean(accuracy_scores):.4f}")
print(f"Precision: {np.mean(precision_scores):.4f}")
print(f"Recall: {np.mean(recall_scores):.4f}")
print(f"F1-Score: {np.mean(f1_scores):.4f}")
print(f"Kappa: {np.mean(kappa_scores):.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat DataFrame dari skor-skor yang telah dihitung
df_baseline = pd.DataFrame({
    'Accuracy': accuracy_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1_Score': f1_scores,
    'Kappa': kappa_scores
})

# Ubah ke bentuk long untuk plotting (melt)
df_melted = df_baseline.melt(var_name='Metrik', value_name='Skor')

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_melted, x='Metrik', y='Skor', palette='Set3')

# Tambahkan judul dan label
plt.title('Distribusi Metrik Evaluasi - Baseline SVM (10-Fold CV)', fontsize=14)
plt.ylabel('Nilai')
plt.xlabel('Metrik')

plt.tight_layout()
plt.show()
```

Simpan Hasil Evaluasi Baseline SVM ke CSV

```python
import pandas as pd

# Buat DataFrame dari hasil evaluasi SVM
baseline_results = pd.DataFrame({
    'Fold': list(range(1, 11)),
    'Accuracy': accuracy_scores,
    'Precision': precision_scores,
    'Recall': recall_scores,
    'F1_Score': f1_scores,
    'Kappa': kappa_scores
})

# Simpan ke Google Drive
output_path = '/content/drive/MyDrive/Colab Notebooks/baseline_svm_results.csv'
baseline_results.to_csv(output_path, index=False)

print(f"Hasil evaluasi baseline SVM berhasil disimpan ke: {output_path}")
```

# Seleksi Fitur SVM

# a. Correlation-based Feature Selection (CFS Approximation)

```python
# Gabungkan kembali X dan y untuk analisis korelasi
df_corr = X_scaled_df.copy()
df_corr['label'] = y

# Hitung korelasi Pearson antara fitur dan label
correlations = df_corr.corr()['label'].drop('label')

# Tampilkan korelasi fitur (dari tinggi ke rendah)
print("Korelasi fitur terhadap label:")
print(correlations.sort_values(ascending=False))

# Threshold korelasi (misal 0.1)
threshold = 0.1
selected_features_cfs = correlations[abs(correlations) > threshold].index.tolist()

# Informasi hasil seleksi
print(f"\nJumlah fitur terpilih (|corr| > {threshold}): {len(selected_features_cfs)}")
print("Fitur terpilih:")
print(selected_features_cfs)

# Buat DataFrame fitur terpilih
X_cfs = X_scaled_df[selected_features_cfs]
```

```python
from sklearn.svm import SVC

# Inisialisasi ulang model SVM dengan kernel RBF
svm_cfs = SVC(kernel='rbf', random_state=42)

# Reset metrik
acc_cfs, prec_cfs, rec_cfs, f1_cfs, kappa_cfs = [], [], [], [], []

for fold, (train_idx, test_idx) in enumerate(skf.split(X_cfs, y), 1):
    X_train, X_test = X_cfs.iloc[train_idx], X_cfs.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    svm_cfs.fit(X_train, y_train)
    y_pred = svm_cfs.predict(X_test)

    acc_cfs.append(accuracy_score(y_test, y_pred))
    prec_cfs.append(precision_score(y_test, y_pred))
    rec_cfs.append(recall_score(y_test, y_pred))
    f1_cfs.append(f1_score(y_test, y_pred))
    kappa_cfs.append(cohen_kappa_score(y_test, y_pred))

    print(f"Fold {fold} selesai (SVM + CFS)")

# Rata-rata hasil
print("\n=== SVM + CFS Evaluation (10-Fold CV) ===")
print(f"Accuracy: {np.mean(acc_cfs):.4f}")
print(f"Precision: {np.mean(prec_cfs):.4f}")
print(f"Recall: {np.mean(rec_cfs):.4f}")
print(f"F1-Score: {np.mean(f1_cfs):.4f}")
print(f"Kappa: {np.mean(kappa_cfs):.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat DataFrame dari hasil evaluasi
df_svm_cfs = pd.DataFrame({
    'Accuracy': acc_cfs,
    'Precision': prec_cfs,
    'Recall': rec_cfs,
    'F1_Score': f1_cfs,
    'Kappa': kappa_cfs
})

# Ubah ke format long (melt) agar bisa digunakan oleh seaborn
df_svm_cfs_melted = df_svm_cfs.melt(var_name='Metrik', value_name='Skor')

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_svm_cfs_melted, x='Metrik', y='Skor', palette='Set2')

# Tambahkan judul dan label sumbu
plt.title('Distribusi Metrik Evaluasi - SVM (CFS)', fontsize=14)
plt.xlabel('Metrik')
plt.ylabel('Skor')

plt.tight_layout()
plt.show()
```

Simpan hasil evaluasi SVM + CFS ke CSV

```python
svm_cfs_results = pd.DataFrame({
    'Fold': list(range(1, 11)),
    'Accuracy': acc_cfs,
    'Precision': prec_cfs,
    'Recall': rec_cfs,
    'F1_Score': f1_cfs,
    'Kappa': kappa_cfs
})

output_path_cfs = '/content/drive/MyDrive/Colab Notebooks/svm_cfs_results.csv'
svm_cfs_results.to_csv(output_path_cfs, index=False)

print(f"Hasil evaluasi SVM + CFS disimpan ke: {output_path_cfs}")
```

# b. Rough Set

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Hitung mutual information setiap fitur terhadap label
mi_scores = mutual_info_classif(X_scaled_df, y, discrete_features=False)
mi_series = pd.Series(mi_scores, index=X_scaled_df.columns).sort_values(ascending=False)

# Pilih Top-N fitur
top_n = 20
selected_features_rs = mi_series.head(top_n).index.tolist()

# Tampilkan hasil seleksi fitur
print(f"Fitur terpilih dari Rough Set Simplified (Top {top_n} berdasarkan Mutual Information):")
print(selected_features_rs)

# Dataset akhir dengan fitur terpilih
X_rs = X_scaled_df[selected_features_rs]
```

```python
from sklearn.svm import SVC

# Inisialisasi model SVM
svm_rs = SVC(kernel='rbf', random_state=42)

# Reset metrik
acc_rs, prec_rs, rec_rs, f1_rs, kappa_rs = [], [], [], [], []

for fold, (train_idx, test_idx) in enumerate(skf.split(X_rs, y), 1):
    X_train, X_test = X_rs.iloc[train_idx], X_rs.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]

    svm_rs.fit(X_train, y_train)
    y_pred = svm_rs.predict(X_test)

    acc_rs.append(accuracy_score(y_test, y_pred))
    prec_rs.append(precision_score(y_test, y_pred))
    rec_rs.append(recall_score(y_test, y_pred))
    f1_rs.append(f1_score(y_test, y_pred))
    kappa_rs.append(cohen_kappa_score(y_test, y_pred))

    print(f"Fold {fold} selesai")

# Rata-rata hasil
print("\n=== SVM + RS Evaluation (10-Fold CV) ===")
print(f"Accuracy: {np.mean(acc_rs):.4f}")
print(f"Precision: {np.mean(prec_rs):.4f}")
print(f"Recall: {np.mean(rec_rs):.4f}")
print(f"F1-Score: {np.mean(f1_rs):.4f}")
print(f"Kappa: {np.mean(kappa_rs):.4f}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Buat DataFrame dari hasil evaluasi
df_svm_rs = pd.DataFrame({
    'Accuracy': acc_rs,
    'Precision': prec_rs,
    'Recall': rec_rs,
    'F1_Score': f1_rs,
    'Kappa': kappa_rs
})

# Ubah ke format long (melt) agar bisa digunakan oleh seaborn
df_svm_rs_melted = df_svm_rs.melt(var_name='Metrik', value_name='Skor')

# Plot boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_svm_rs_melted, x='Metrik', y='Skor', palette='Set2')

# Tambahkan judul dan label sumbu
plt.title('Distribusi Metrik Evaluasi - SVM (RS)', fontsize=14)
plt.xlabel('Metrik')
plt.ylabel('Skor')

plt.tight_layout()
plt.show()
```
